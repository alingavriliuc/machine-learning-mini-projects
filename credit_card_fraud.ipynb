```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
dataset_dir = "./datasets/"
credit_card_dataset = pd.read_csv(dataset_dir + "creditcard.csv")
```

```python
# balancing 
legit = credit_card_dataset[credit_card_dataset.Class == 0]
fraud = credit_card_dataset[credit_card_dataset.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_dataset.groupby('Class').mean()
```

```python
# Under sampling
legit_sample = legit.sample(n=492)
new_dataset = pd.concat([legit_sample, fraud], axis=0)
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
# en gros ce qui nous avons fait c'est de prendre 492 transactions legit et 492 transactions fraud
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(train_prediction, Y_train)
print("Accuracy on training data : ", training_data_accuracy)
```

```python
test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(test_prediction, Y_test)
print("Accuracy on test data : ", test_data_accuracy)
```
